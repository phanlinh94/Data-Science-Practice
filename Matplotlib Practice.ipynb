```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

# Lineplot

```python
x = np.linspace(0, 10, 1000)
ax = plt.axes()
ax.plot(x, np.sin(x), '--g')
ax.plot(x, np.sin(x-1), '-b')
ax.set(xlim=(0, 10), ylim=(-2, 2), title = "A Simple Graph", xlabel = "x", ylabel = "sin(x)")
```

# Scatterplot

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
plt.scatter(x, y, c=colors, s=100, alpha=0.3, cmap='viridis')
plt.colorbar()
```

For bigger dataset, `plt.plot` is more efficient

# Basic errorbars

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k', color='black', ecolor='orange', capsize=10)
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='gray', ecolor='lightgray', capsize=2)
```

# Density & Contour plots

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y*x) * np.cos(x)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

`np.meshgrid` function builds two-dimensional grids from one-dimensional arrays.

```python
plt.contour(X, Y, Z, colors='black')
```

*negative values are represented by dashed lines*

```python
plt.contourf(X, Y, Z, 20, cmap="RdGy")
plt.colorbar()
```

*here the **colormap is Red-Gray**, gap between lines are filled by using `plt.contourf()` function*

```python
contours = plt.contour(X, Y, Z, 3, colors = 'black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin = 'lower', cmap = 'RdGy', alpha = 0.5)
plt.colorbar()
```

# Customize the plot legends

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label = 'sine')
ax.plot(x, np.cos(x), '--g', label = "cosine")
ax.axis('equal')
leg = ax.legend(loc='upper right', #location of the legend
                frameon=True, #legend frame or not
                ncol=2, #number of columns
                fancybox=True, #rounded legend box
                framealpha=0.3, #opacity
                shadow=True, #shadow around the legend
                borderpad=1) #border
```

```python
# Extract the data from CSV
age, bmi = stroke['age'], stroke['bmi']
stroked, av_glu = stroke['stroke'], stroke['avg_glucose_level']

#Scatter the points, using size and color but no label
plt.scatter(age, bmi, label = None,
           c = stroked, cmap='viridis',
           s = av_glu, linewidth=0, alpha = 0.3)
plt.axis('equal')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.colorbar(label='Had a stroke')

# Here we create a legend:
# we'll plot empty lists with the desired size and label
for av_glu in [100, 200, 300]:
    plt.scatter([], [], c='purple', alpha = 0.3, s = av_glu,
               label = str(av_glu) + ' mmol/L')
plt.legend(scatterpoints=1, frameon=True, labelspacing=1, title = "Average Glucose Level")
plt.figure(figsize=(100, 70))
plt.show()
```

# Customize colorbar

```python
#Scatter the points, using size and color but no label
plt.scatter(age, bmi, label = None,
           c = stroked, cmap=plt.cm.get_cmap('Blues', 2),
           s = av_glu, linewidth=0, alpha = 0.1)
plt.axis('equal')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.colorbar(label='Had a stroke')

# Here we create a legend:
# we'll plot empty lists with the desired size and label
for av_glu in [100, 200, 300]:
    plt.scatter([], [], c='purple', alpha = 0.3, s = av_glu,
               label = str(av_glu) + ' mmol/L')
plt.legend(scatterpoints=1, frameon=True, labelspacing=1, title = "Average Glucose Level")
```
